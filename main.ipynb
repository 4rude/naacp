## What can we learn from this?
Having low prison populations of any demographic is generally considered a good thing, and women are incarcerated at significantly lower rates (making up roughly 7% of the US prison population). However, one must be careful not to overlook problems specific to imprisoned women in Minnesota. For instance, women prisoners deal with the unique issues of childbearing while incarcerated. Another issue specific to Minnesota is that Native American women are imprisoned at a much higher rate than other demographics (Mitchell, 2017). These issues highlight the need to carefully analyse data to ensure that large incarceration disparities like this do not go unnoticed.
